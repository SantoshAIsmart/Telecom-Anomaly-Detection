```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. ‡§™‡•ç‡§∞‡•ã‡§´‡•á‡§∂‡§®‡§≤ ‡§¨‡§ø‡§π‡§æ‡§∞ ‡§°‡•á‡§ü‡§æ (Phase 2 ‡§ï‡•á ‡§≤‡§ø‡§è)
data = {
    'Site_Name': ['Patna_Central', 'Muzaffarpur_Main', 'Gaya_Tower_1', 'Bhagalpur_Industrial', 'Darbhanga_City', 'Siwan_Rural', 'Motihari_Market'],
    'Latency_ms': [22, 55, 30, 48, 25, 65, 38],
    'Packet_Loss_Pct': [0.05, 2.1, 0.2, 1.8, 0.1, 3.5, 0.5], # Puch.ai ‡§ï‡§æ ‡§∏‡•Å‡§ù‡§æ‡§µ #1
    'Uptime_Pct': [99.9, 94.5, 99.1, 95.2, 99.8, 91.2, 98.5], # Puch.ai ‡§ï‡§æ ‡§∏‡•Å‡§ù‡§æ‡§µ #2
    'User_Load': [1200, 3500, 1800, 2900, 1400, 800, 2100]
}

df_phase2 = pd.DataFrame(data)

# 2. ‡§Æ‡§≤‡•ç‡§ü‡•Ä-‡§ó‡•ç‡§∞‡§æ‡§´ ‡§è‡§®‡§æ‡§≤‡§ø‡§∏‡§ø‡§∏ (Professional Dashboard)
fig, ax1 = plt.subplots(figsize=(12,6))

# Latency ‡§ï‡§æ ‡§¨‡§æ‡§∞ ‡§ö‡§æ‡§∞‡•ç‡§ü
ax1.bar(df_phase2['Site_Name'], df_phase2['Latency_ms'], color='skyblue', label='Latency (ms)')
ax1.set_ylabel('Latency (ms)', color='blue')

# Packet Loss ‡§ï‡•Ä ‡§≤‡§æ‡§á‡§® (Secondary Axis)
ax2 = ax1.twinx()
ax2.plot(df_phase2['Site_Name'], df_phase2['Packet_Loss_Pct'], color='red', marker='D', linewidth=2, label='Packet Loss %')
ax2.set_ylabel('Packet Loss (%)', color='red')

plt.title('Phase 2: Network Health Analysis (Bihar Sites)', fontsize=15)
ax1.set_xticklabels(df_phase2['Site_Name'], rotation=45)
plt.grid(alpha=0.3)
plt.show()

print("‚úÖ Phase 2 Model Ready: Latency aur Packet Loss ka correlation dikh raha hai!")
```

```python
# ‡§Ö‡§ó‡§≤‡•á 24 ‡§ò‡§Ç‡§ü‡•ã‡§Ç ‡§ï‡•á ‡§≤‡§ø‡§è ‡§®‡•á‡§ü‡§µ‡§∞‡•ç‡§ï ‡§π‡•á‡§≤‡•ç‡§• ‡§ï‡•Ä ‡§≠‡§µ‡§ø‡§∑‡•ç‡§Ø‡§µ‡§æ‡§£‡•Ä
import numpy as np

# ‡§®‡§ï‡§≤‡•Ä 24 ‡§ò‡§Ç‡§ü‡•á ‡§ï‡§æ ‡§°‡•á‡§ü‡§æ ‡§¨‡§®‡§æ‡§®‡§æ
hours = np.arange(24).reshape(-1, 1)
baseline_latency = 30 + 15 * np.sin(hours * (2 * np.pi / 24)) + np.random.normal(0, 2, hours.shape)

# AI ‡§ï‡•ã ‡§≠‡§µ‡§ø‡§∑‡•ç‡§Ø‡§µ‡§æ‡§£‡•Ä ‡§ï‡§∞‡§®‡§æ ‡§∏‡§ø‡§ñ‡§æ‡§®‡§æ
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(hours, baseline_latency)

# ‡§Ö‡§ó‡§≤‡•á 6 ‡§ò‡§Ç‡§ü‡•ã‡§Ç ‡§ï‡§æ ‡§Ö‡§®‡•Å‡§Æ‡§æ‡§®
future_hours = np.arange(24, 30).reshape(-1, 1)
prediction = model.predict(future_hours)

plt.figure(figsize=(10,4))
plt.plot(hours, baseline_latency, label='Past Data (24h)')
plt.plot(future_hours, prediction, 'r--', label='AI Prediction (Next 6h)')
plt.title('Santosh Kumar: Network Latency Forecast')
plt.xlabel('Hour of the Day')
plt.ylabel('Latency (ms)')
plt.legend()
plt.grid(True)
plt.show()

print("üéØ Prediction Complete: Agle 6 ghanto ka anuman tayyar hai!")
```
